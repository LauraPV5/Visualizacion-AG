```python
#Importamos modulos
import pandas as pd
import io
import requests
import seaborn as sns
import timeit
import matplotlib.pyplot as plt
```

# Primero adquirimos los datos

```python
#Cargamos el data set de los pasajeros del Titanic
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
s=requests.get(url).content
titanic=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
#Informacion del dataset
titanic.info()
```

Exploramos los datos

```python
#Renombra la columna class=clasey fare=tarifa
titanic.rename(columns={'class': 'clase'}, 
inplace=True)
titanic.rename(columns={'fare': 'tarifa'}, 
inplace=True)
#Muestra los valores distintos para class(clase)
titanic.clase.unique()
```

```python
#Primeras 5 filas
titanic.head(5)
```

Definimos nuevas columnas en el dataset

```python
#Añadimos nuevas columnas: is_old, is_baby
def is_old_func(row):
  return row['age'] > 60
titanic['is_old'] = titanic.apply(is_old_func, axis='columns')

titanic.eval ('is_baby = age < 15', inplace = True)
titanic.head(5)
```

```python
#Define una variable numérica: class_num
def class_num_func(row):
  clase = {'Third':3,'First':1,'Second':2}
  return clase[row.clase]
titanic['calss_num'] = titanic.apply(class_num_func, axis='columns')
#Primeras 5 filas
titanic.head(5)
```

# Consulta de datos con condiciones

```python
#Consulta con condiciones: las mujeres que son de primera o tercera clase y las mayores de 45 años que no supervivieron
titanic[
    (titanic.sex == 'female')
  & (titanic['clase'].isin(['First', 'Third']))
  & (titanic.age > 45)
  & (titanic.survived == 0)
]
```

Distribución de las clases

```python
#Distribución de los pasajeros por clase
sns.countplot(x='pclass', data=titanic)
```

Distribución de la edad

```python
#Distribución de los pasajeros por edad(ege)
sns.distplot(titanic.age.dropna())
plt.show()

sns.histplot(titanic.age.dropna())
plt.show()
```

BoxPlot de la edad por sexo y clase

```python
#Boxplot de la edad por sexo y clase
with sns.axes_style(style='ticks'):
  ax=sns.catplot('sex','age','clase',data=titanic, kind='box')
```

Distribución cruzada de edad y tarifa

```python
#Distribución cruzada Edad y Tarifa
sns.jointplot(x='tarifa', y='age', data=titanic)
```

Distribución de la edad según las categorías: survived, clase

```python
#Cambiamos el font
sns.set(font_scale=1)

#FacetGrid - Contrsuir una matriz de gráficos
g= sns.FacetGrid(titanic,row='survived',col='clase')
g.map(sns.distplot,'age')
plt.show() 
```

Distribución de la edad según las categorías: survived, clase

```python
#Diagra de dispersion con distribución de cada variable: fare(precio)/age(edad)
sns.jointplot(data=titanic,x='age',y='tarifa',kind='reg',color='b')
plt.show()
```

```python
#Mapa de calor de correlaciones
tc=titanic.corr()
sns.heatmap(tc,cmap='coolwarm')
plt.title('titanic.corr()')
```

Diagramas de dispersión de parejas de variables cuantitativas

```python
##Define un subconjuto de datos con las variables numéricas
titanic_num = titanic[['survived','pclass','sibsp','parch','tarifa']]
#Hace una matriz de diagramas de dispersión de parejas de variables.
sns.pairplot(titanic_num, hue="survived")
plt.show()
```
